```python
import numpy as np


prem_smoker = 19
total_smoker = 126

prem_non = 133
total_non = 873


p1 = prem_smoker / total_smoker
p2 = prem_non / total_non

se = np.sqrt(p1*(1-p1)/total_smoker + p2*(1-p2)/total_non)


lower = (p1 - p2) - 1.96*se
upper = (p1 - p2) + 1.96*se

(lower, upper)
```

```python
import statsmodels.api as sm

count = np.array([19, 133])
nobs = np.array([126, 873])

z_stat, p_value = sm.stats.proportions_ztest(count, nobs)
z_stat, p_value
```
